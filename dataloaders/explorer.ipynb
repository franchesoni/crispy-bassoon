```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import json

truckindex = 574
filedir = datapath.parent / index_ade20k['folder'][truckindex]
filename = index_ade20k['filename'][truckindex]
filenames = [f for f in os.listdir(filedir) if filename.split('.')[0].split('_')[-1] in f]
for f in filenames:
    if f.endswith('png') or f.endswith('jpg'):
        plt.figure()
        plt.imshow(Image.open(filedir / f))
        plt.axis('off')
        plt.title(f)

with open(
filedir / ((filedir / filename).stem + '.json')
) as f:
    ann = json.load(f)
truck_masks = [obj['instance_mask'] for obj in ann['annotation']['object'] if 'truck' in obj['name']]
for truckmask in truck_masks:
    plt.figure()
    plt.imshow(Image.open(filedir / truckmask))
    plt.axis('off')
    plt.title(truckmask)




```
